```python
mean_value = df['DewPointHighF'].mean()
df['DewPointHighF'] = df['DewPointHighF'].fillna(mean_value)

mean_value = df['DewPointAvgF'].mean()
df['DewPointAvgF'] = df['DewPointAvgF'].fillna(mean_value)

mean_value = df['DewPointLowF'].mean()
df['DewPointLowF'] = df['DewPointLowF'].fillna(mean_value)

mean_value = df['HumidityHighPercent'].mean()
df['HumidityHighPercent'] = df['HumidityHighPercent'].fillna(mean_value)

mean_value = df['HumidityAvgPercent'].mean()
df['HumidityAvgPercent'] = df['HumidityAvgPercent'].fillna(mean_value)

mean_value = df['HumidityLowPercent'].mean()
df['HumidityLowPercent'] = df['HumidityLowPercent'].fillna(mean_value)

mean_value = df['SeaLevelPressureHighInches'].mean()
df['SeaLevelPressureHighInches'] = df['SeaLevelPressureHighInches'].fillna(mean_value)

mean_value = df['SeaLevelPressureAvgInches'].mean()
df['SeaLevelPressureAvgInches'] = df['SeaLevelPressureAvgInches'].fillna(mean_value)

mean_value = df['SeaLevelPressureLowInches'].mean()
df['SeaLevelPressureLowInches'] = df['SeaLevelPressureLowInches'].fillna(mean_value)

mean_value = df['VisibilityHighMiles'].mean()
df['VisibilityHighMiles'] = df['VisibilityHighMiles'].fillna(mean_value)

mean_value = df['VisibilityAvgMiles'].mean()
df['VisibilityAvgMiles'] = df['VisibilityAvgMiles'].fillna(mean_value)

mean_value = df['VisibilityLowMiles'].mean()
df['VisibilityLowMiles'] = df['VisibilityLowMiles'].fillna(mean_value)

mean_value = df['WindHighMPH'].mean()
df['WindHighMPH'] = df['WindHighMPH'].fillna(mean_value)

mean_value = df['WindAvgMPH'].mean()
df['WindAvgMPH'] = df['WindAvgMPH'].fillna(mean_value)

mean_value = df['WindGustMPH'].mean()
df['WindGustMPH'] = df['WindGustMPH'].fillna(mean_value)

```

```python
# 2. Outlier Detection and Removal:
```

```python
df
```

```python

# List of columns for outlier detection
cols = [
    'DewPointHighF', 'DewPointAvgF', 'DewPointLowF',
    'HumidityHighPercent', 'HumidityAvgPercent', 'HumidityLowPercent',
    'SeaLevelPressureHighInches', 'SeaLevelPressureAvgInches', 'SeaLevelPressureLowInches',
    'VisibilityHighMiles', 'VisibilityAvgMiles', 'VisibilityLowMiles',
    'WindHighMPH', 'WindAvgMPH', 'WindGustMPH'
]

# Ensure columns are numeric
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Store indices of outliers for reporting
outlier_indices = set()

# Detect and remove outliers column by column
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}:")
    print(f"  Q1  = {Q1}")
    print(f"  Q3  = {Q3}")
    print(f"  IQR = {IQR}")
    print(f"  Outliers found: {outliers.shape[0]}\n")
    outlier_indices.update(outliers.index.tolist())

# Remove all rows that are outliers in any of the columns
df_clean = df.drop(index=outlier_indices)

print(f"Total rows removed: {len(outlier_indices)}")
print(f"Remaining rows: {df_clean.shape[0]}")
```

```python
from sklearn.preprocessing import MinMaxScaler
# Convert non-numeric values (like 'T') to NaN, then convert all to numeric
numeric_cols = [
    'TempHighF', 'TempLowF', 'DewPointHighF', 'DewPointLowF',
    'HumidityHighPercent', 'HumidityLowPercent',
    'SeaLevelPressureHighInches', 'SeaLevelPressureLowInches',
    'VisibilityHighMiles', 'VisibilityLowMiles',
    'WindHighMPH', 'WindGustMPH'
]

# Convert columns to numeric (if not already)
for col in numeric_cols:
    df_reduced[col] = pd.to_numeric(df_reduced[col], errors='coerce')

# Apply Min-Max Scaling
scaler = MinMaxScaler()
df_scaled = df_reduced.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df_reduced[numeric_cols])

# Show the scaled data
df_scaled[numeric_cols]
```

```python
# save the scaled data into a csv file
df_scaled.to_csv('min_max_scaled_austin_weather.csv', index=False)
```

```python
# Discretization (Binning into Low, Medium, High)
columns_to_bin = ['TempHighF', 'TempLowF', 'DewPointHighF',
                  'HumidityHighPercent', 'VisibilityLowMiles', 'WindHighMPH']

for col in columns_to_bin:
    binned_col = col + "_binned"
    df_scaled[binned_col] = pd.cut(df_scaled[col], bins=3, labels=["Low", "Medium", "High"])

# Show the discretized columns
print(df_scaled[[col + "_binned" for col in columns_to_bin]].head())

# Optionally, save the updated DataFrame to a new CSV
df_scaled.to_csv("austin_weather_scaled_binned_after_Discretization.csv", index=False)
```
